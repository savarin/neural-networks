```python
import numpy as np
import pandas as pd
from time import time
from sklearn.preprocessing import StandardScaler

np.random.seed(1337)

df = pd.read_csv('mnist.csv')
```

```python
df_train = df.iloc[:33600, :]

X_train = df_train.iloc[:, 1:].values / 255.
y_train = pd.get_dummies(df_train['label']).values
```

```python
df_test = df.iloc[33600:, :]

X_test = df_test.iloc[:, 1:].values / 255.
y_test = pd.get_dummies(df_test['label']).values
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=0)
model = model.fit(X_train, df_train['label'].values)

y_prediction = model.predict(X_test)
print "accuracy", np.sum(y_prediction == df_test['label'].values) / float(len(y_test))
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.imshow(X_train[0].reshape(28,28), cmap=plt.get_cmap('gray'))
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation

model = Sequential()
model.add(Dense(input_dim=784, output_dim=10))
model.add(Activation("softmax"))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train)
```

```python
model.evaluate(X_test, y_test)
```
